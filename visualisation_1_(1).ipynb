Exercise 1: Simple Polynomial Plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Création d'un tableau x avec 100 points entre -10 et 10
x = np.linspace(-10, 10, 100)

# Calcul de la fonction polynomiale y = 2x^3 - 5x^2 + 3x - 7
y = 2*x**3 - 5*x**2 + 3*x - 7

# Définition de la figure avec une taille de 10x6 pouces
plt.figure(figsize=(10, 6), dpi=80)

# Création d'un sous-graphique unique
plt.subplot(111)

# Tracé de la courbe en bleu avec une ligne continue de largeur 2.5
plt.plot(x, y, color="blue", linewidth=2.5, linestyle="-")

# Ajout des labels des axes
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

# Ajout du titre du graphique
plt.title(r"Tracé de la fonction polynomiale $y = 2x^3 - 5x^2 + 3x - 7$", fontsize=16)

# Ajout de la grille
plt.grid(True)

# Affichage du graphique
plt.show()
```

Exercise 2: Exponential and Logarithmic Plot


```python
import numpy as np
import matplotlib.pyplot as plt

# Création d'un tableau x avec 500 points entre 0.1 et 10
x = np.linspace(0.1, 10, 500)

# Calcul des fonctions exponentielle et logarithmique
y1 = np.exp(x)  # exponentielle
y2 = np.log(x)  # logarithme

# Définition de la figure avec une taille de 10x6 pouces et 100 DPI
plt.figure(figsize=(10, 6), dpi=100)

# Création d'un sous-graphique unique
plt.subplot(111)

# Tracé des courbes avec des styles différents
plt.plot(x, y1, color="blue", linewidth=2.5, linestyle="-", label=r"$y = e^x$")
plt.plot(x, y2, color="red", linewidth=2.5, linestyle="--", label=r"$y = \ln(x)$")

# Ajout des labels des axes
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

# Ajout du titre du graphique
plt.title(r"Tracé des fonctions $y = e^x$ et $y = \ln(x)$", fontsize=16)

# Ajout de la grille
plt.grid(True)

# Ajout de la légende
plt.legend(loc="upper left", frameon=False)

# Sauvegarde de l'image en PNG avec 100 DPI
plt.savefig("exponential_logarithm_plot.png", format='png', dpi=100)

# Affichage du graphique
plt.show()
```

Exercise 3: Figure and Subplots

```python
# Création d'un tableau x pour les différentes fonctions
x1 = np.linspace(-2*np.pi, 2*np.pi, 500)  # Pour tan et arctan
x2 = np.linspace(-2, 2, 500)  # Pour sinh et cosh

# Calcul des fonctions
y_tan = np.tan(x1)
y_arctan = np.arctan(x1)

y_sinh = np.sinh(x2)
y_cosh = np.cosh(x2)

# Création de la figure avec deux sous-graphiques
fig, axes = plt.subplots(1, 2, figsize=(12, 5), dpi=100)

# Premier subplot : tan et arctan
axes[0].plot(x1, y_tan, color="blue", linewidth=2, linestyle="-", label=r"$y = \tan(x)$")
axes[0].plot(x1, y_arctan, color="red", linewidth=2, linestyle="--", label=r"$y = \arctan(x)$")
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title(r"Fonctions $\tan(x)$ et $\arctan(x)$")
axes[0].legend()
axes[0].grid(True)
axes[0].set_ylim(-10, 10)  # Limite de l'axe y pour éviter les valeurs infinies de tan(x)

# Deuxième subplot : sinh et cosh
axes[1].plot(x2, y_sinh, color="green", linewidth=2, linestyle="-", label=r"$y = \sinh(x)$")
axes[1].plot(x2, y_cosh, color="purple", linewidth=2, linestyle="--", label=r"$y = \cosh(x)$")
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title(r"Fonctions $\sinh(x)$ et $\cosh(x)$")
axes[1].legend()
axes[1].grid(True)

# Ajustement de la mise en page
fig.tight_layout()

# Affichage du graphique
plt.show()




```


Exercise 5: Scatter Plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Génération des données
np.random.seed(42)  # Pour la reproductibilité
x = np.linspace(0, 10, 500)  # Valeurs uniformément réparties
y = np.sin(x) + np.random.normal(0, 0.2, 500)  # Fonction sin(x) avec bruit

# Création des tailles et couleurs des marqueurs en fonction des valeurs de y
marker_sizes = 50 * (np.abs(y) + 0.2)  # Ajustement de la taille des marqueurs
colors = y  # Couleurs basées sur les valeurs de y

# Définition de la figure avec une taille de 10x6 pouces et 100 DPI
plt.figure(figsize=(10, 6), dpi=100)

# Tracé du nuage de points (scatter plot)
plt.scatter(x, y, s=marker_sizes, c=colors, cmap='viridis', alpha=0.75, edgecolors="black")

# Ajout des labels et du titre
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.title("Nuage de points : $y = \sin(x) + \text{bruit}$", fontsize=16)

# Ajout de la grille
plt.grid(True)

# Suppression des ticks des axes
plt.xticks([])
plt.yticks([])

# Sauvegarde du graphique au format PDF
plt.savefig("scatter_plot.pdf", format='pdf', dpi=100)

# Affichage du graphique
plt.show()
```


Exercise 6: Contour Plot

```python
# Création des tableaux x et y avec 200 points entre -5 et 5
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)

# Création de la grille 2D
X, Y = np.meshgrid(x, y)

# Définition de la fonction f(x, y) = sin(sqrt(x^2 + y^2))
Z = np.sin(np.sqrt(X**2 + Y**2))

# Définition de la figure
plt.figure(figsize=(8, 6), dpi=100)

# Tracé des contours remplis avec un colormap différent (plasma)
contour_filled = plt.contourf(X, Y, Z, levels=20, cmap="plasma", alpha=0.75)

# Tracé des contours avec lignes noires
contour_lines = plt.contour(X, Y, Z, levels=20, colors="black")

# Ajout des labels sur les contours
plt.clabel(contour_lines, inline=True, fontsize=10)

# Ajout des labels et du titre
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)
plt.title("Tracé des contours de la fonction $f(x, y) = \sin(\sqrt{x^2 + y^2})$", fontsize=14)

# Ajout d'une barre de couleur
plt.colorbar(contour_filled)

# Affichage du graphique
plt.show()
```


Exercise 7: 3D Surface Plot

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Création des tableaux X et Y avec un pas de 0.25 entre -5 et 5
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(x, y)

# Calcul de la fonction Z = cos(sqrt(X^2 + Y^2))
Z = np.cos(np.sqrt(X**2 + Y**2))

# Création de la figure et du subplot 3D
fig = plt.figure(figsize=(10, 6), dpi=100)
ax = fig.add_subplot(111, projection='3d')

# Tracé de la surface avec une colormap différente (coolwarm)
surface = ax.plot_surface(X, Y, Z, cmap="coolwarm", linewidth=0, antialiased=False)

# Ajout d'une barre de couleur
fig.colorbar(surface, shrink=0.5, aspect=10)

# Changement de l'angle de vue (élévation, azimut)
ax.view_init(elev=30, azim=45)

# Ajout des labels
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Surface 3D de $Z = \cos(\sqrt{X^2 + Y^2})$")

# Affichage du graphique
plt.show()
```


Exercise 9: Logarithmic Scale

```python
import numpy as np
import matplotlib.pyplot as plt

# Création de l'array x avec 50 points entre 1 et 100
x = np.linspace(1, 100, 50)

# Calcul des fonctions
y1 = 2**x  # Exponentielle base 2
y2 = np.log2(x)  # Logarithme base 2

# Définition de la figure avec une taille de 12x6 pouces
plt.figure(figsize=(12, 6), dpi=100)

# Tracé des courbes avec échelle logarithmique sur l'axe y
plt.plot(x, y1, color="blue", linestyle="-", linewidth=2, label=r"$y = 2^x$")
plt.plot(x, y2, color="red", linestyle="--", linewidth=2, label=r"$y = \log_2(x)$")

# Mise en échelle logarithmique de l'axe y
plt.yscale("log")

# Ajout des labels et du titre
plt.xlabel("x", fontsize=14)
plt.ylabel("y (échelle logarithmique)", fontsize=14)
plt.title("Tracé des fonctions $y = 2^x$ et $y = \log_2(x)$ avec échelle logarithmique", fontsize=16)

# Ajout de la grille
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

# Ajout d'une légende
plt.legend()

# Affichage du graphique
plt.show()
```

Exercise 12: 3D Contour Plot

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Création des tableaux X et Y avec un pas de 0.25 entre -5 et 5
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(x, y)

# Calcul de la fonction Z = exp(-0.1 * (X^2 + Y^2))
Z = np.exp(-0.1 * (X**2 + Y**2))

# Création de la figure et du subplot 3D
fig = plt.figure(figsize=(10, 6), dpi=100)
ax = fig.add_subplot(111, projection='3d')

# Tracé des contours 3D avec une colormap différente
contour = ax.contour3D(X, Y, Z, 50, cmap="inferno")

# Ajout des labels des axes
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Contours 3D de $Z = e^{-0.1 (X^2 + Y^2)}$")

# Ajout d'une barre de couleur
fig.colorbar(contour, shrink=0.5, aspect=10)

# Affichage du graphique
plt.show()
```

Exercise 14: 3D Bar Plot

```python
# Création des tableaux x et y avec 10 points entre -5 et 5
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Aplatir les tableaux pour une utilisation avec bar3d
X_flat = X.flatten()
Y_flat = Y.flatten()

# Calcul des valeurs de Z
Z = np.exp(-0.1 * (X_flat**2 + Y_flat**2))

# Largeur et profondeur des barres
dx = dy = 0.5  # Taille des barres

# Création de la figure et du subplot 3D
fig = plt.figure(figsize=(10, 6), dpi=100)
ax = fig.add_subplot(111, projection='3d')

# Définition de la normalisation des couleurs
norm = plt.Normalize(vmin=Z.min(), vmax=Z.max())
colors = plt.cm.coolwarm(norm(Z))  # Application de la colormap

# Tracé du bar plot en 3D avec couleurs adaptées aux valeurs de Z
bars = ax.bar3d(X_flat, Y_flat, np.zeros_like(Z), dx, dy, Z, color=colors, shade=True)

# Création d'un ScalarMappable pour la barre de couleur
sm = plt.cm.ScalarMappable(cmap="coolwarm", norm=norm)
sm.set_array([])  # Nécessaire pour éviter l'erreur

# Ajout d'une barre de couleur avec l'axe spécifié
cbar = fig.colorbar(sm, ax=ax, shrink=0.5, aspect=10, pad=0.1)
cbar.set_label("Valeurs de Z")

# Changement de l'angle de vue
ax.view_init(elev=30, azim=45)

# Ajout des labels des axes
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Graphique en barres 3D de $Z = e^{-0.1 (X^2 + Y^2)}$")

# Affichage du graphique
plt.show()
```


Exercise 15: 3D Vector Field

```python
# Création des tableaux X, Y et Z avec des valeurs entre -5 et 5
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
z = np.linspace(-5, 5, 5)
X, Y, Z = np.meshgrid(x, y, z)

# Calcul des composantes du champ vectoriel
U = -Y  # Composante en X
V = X   # Composante en Y
W = Z   # Composante en Z

# Création de la figure et du subplot 3D
fig = plt.figure(figsize=(10, 6), dpi=100)
ax = fig.add_subplot(111, projection='3d')

# Tracé du champ vectoriel avec une couleur et une échelle ajustée
ax.quiver(X, Y, Z, U, V, W, length=0.5, color='red', normalize=True)

# Ajout des labels des axes et du titre
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Champ vectoriel 3D")

# Affichage du graphique
plt.show()
```
